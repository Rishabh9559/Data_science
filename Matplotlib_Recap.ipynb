## **Pie chart**



A pie chart in Matplotlib is best used when you need to visualize proportional data and compare how each category contributes to a meaningful whole.

**When to Use a Matplotlib Pie Chart**
* Displaying percentages
* Highlighting a dominant category
* Limited number of categories
* Comparing parts of a whole

```python
# count number of repeted value
count = df["dept"].value_counts()
count
```

```python
# pie chart
# count
# labels
plt.pie(count, labels = count.index, autopct = "%1.1f" )
plt.show()
```

```python
# explode -> sperated from pie
# explode=[0,0,0...n] labels
plt.pie(count, labels=count.index, autopct="%1.2f", explode=[0,.1,0])
plt.show()
```

## **CountPlot**

* A count plot is used when you need to visualize the frequency of different groups within a single categorical variable.
* It is essentially a bar chart where the height of each bar represents the number of observations (count) for each category.

* **Analyzing categorical data**: Your data has distinct, non-numeric categories (e.g., "product type," "city," "satisfaction rating," or "gender") and you want to understand how many times each category appears.
* **Comparing group sizes**: You want to quickly compare the number of observations across different groups to see which are most or least common.
* **Checking data distribution**: It helps in understanding the distribution of a categorical feature in your dataset, which is a common first step in data analysis and machine learning workflows.

```python
 # countplot

 plt.bar(count.index, count, color="green")
 plt.show()
```

```python
# countplot
# different color for bar
plt.bar(count.index, count, color=["pink","orange","yellow"])
plt.show()
```

## **Scatter plot**

A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis.


**When to Use a Matplotlib Scatter Plot**:

* **Showing Relationships/Correlations**: Scatter plots are ideal for visualizing the relationship between two numerical variables. You can observe if there's a positive correlation (as one variable increases, the other tends to increase), a negative correlation (as one increases, the other tends to decrease), or no clear correlation.
* **Identifying Outliers**: Points that fall far away from the main cluster of data points can be easily identified as outliers.
* **Clustering**: If the data naturally forms distinct groups, a scatter plot can help visualize these clusters.
Comparing Distributions: While not their primary use, scatter plots can give a sense of the distribution of each variable.
* **Visualizing Trends (with a trend line)**: You can often add a regression or trend line to a scatter plot to summarize the relationship between the variables.

## **Pie chart**

```python
# pie chart
# grooup by -> sum
salary_by_dept = df.groupby("dept")["Salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels = salary_by_dept.index, autopct="%1.2f")
plt.show()
```

```python
plt.pie(salary_by_dept, labels=salary_by_dept.index, explode=[0,.1,0], autopct="%1.2f", shadow=True)
plt.show()
```

```python
# Bar Plot

hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
fin_mean = sum(fin_sal)/len(fin_sal)
```

```python
plt.bar(["HR", "IT", "Finance"], [hr_mean,it_mean,fin_mean], color=["green", "pink", "yellow"])
plt.grid()
plt.show()

```

# **Multivariate Analysis**

Multivariate analysis refers to a set of statistical techniques used to analyze data that contains more than two variables.

The primary goal is to understand the relationships between multiple variables simultaneously, rather than examining each variable in isolation.

```python
df["experience"] =  [1, 2, 4, 1.5, 5, 8, 12, 10, 4, 1, 1.2, 2, 3, 1.8, 5.5, 4.5, 3.2, 7, 8, 10]
```

```python
df.head()
```

```python
# Bubble PLot:

plt.scatter(df["Age"], df["Salary"], s = df["experience"]*50, color = "skyblue", edgecolors= "black")
plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# 2 numerical and 1 categorical column:

plt.scatter(df["Age"], df["Salary"], c = df["dept"].map({"HR": "yellow", "IT":"blue", "Finance": "orange"}))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary vs Dept")
plt.legend()
plt.show()
```

```python
color = {"HR": "yellow", "IT":"blue", "Finance": "orange"}

for dept, color in color.items():
    df_dept = df[df["dept"] == dept]
    plt.scatter(df_dept["Age"], df_dept["Salary"], c = color, label = dept)

plt.legend()
```

# **Object Oriented API**

```python
fig, axs = plt.subplots(1,3, figsize = (15,5))

# Line plot

axs[0].plot(sort_age["Age"], df["Salary"], color = "red", marker = "*", linewidth = "2", markersize = 2)
axs[0].grid()
axs[0].set_title("Line Plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

# histogram:

axs[1].hist(df["Salary"], bins = 5, color = "skyblue")
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")


# Boxplot:

axs[2].boxplot(df["Salary"])
axs[2].set_title("Boxplot")
axs[2].set_xlabel("Salary")

# save images
plt.savefig("multipleplots.png")
plt.show()
```
