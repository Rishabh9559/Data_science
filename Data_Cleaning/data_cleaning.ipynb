### **2. Handling Missing values**

* Delete Record (more than **50%** of data is missing)
* Imputation (Filling missing value)
  * **Mean** (distric values)
  * **Median** (Outliers)
  * **Mode** (categorial,not founds)
  * (only for numberical values)
  * **Time series** (Forwords fill / Backwords fill) e.g. date Time, numberical values
  * **Predictive Imputation by using Machine Learning models**



**Remove Missing Primary key -> Customer ID**

```python
# Remove Missing Primary key -> Customer ID
data.dropna(subset=['Customer_ID'], inplace=True)
data
```

```python
data.isnull().sum()
```

**Imputation**

```python
# Imputation on
# Age column

data['Age'].unique()
```

```python
# remove string from years
data['Age'] = data['Age'].str.replace('years','')
```

```python
data['Age'].unique()
```

```python
# remove nan value by median
# remove nan and nana

age_data = data[data['Age'] != 'nan ']['Age']

age_data
```

```python
# dropna and convert into int data types
# Convert the 'Age' column to numeric, coercing any non-numeric values to NaN.
# Then drop the NaN values, and finally convert to integer before calculating the median.
age_median=pd.to_numeric(age_data, errors='coerce').dropna().astype(int).median()
age_median = int(age_median)
age_median
```

```python
# replcae nan with medican of age
data['Age'].replace('nan ', age_median, inplace=True)
```

```python
# replca null value  by median
data['Age'].fillna(age_median, inplace=True)
```

```python
data['Age'].unique()
```

```python
# Purchase amount -> replace null value by median

data['Purchase_Amount'].fillna(data['Purchase_Amount'].median(), inplace=True)

```

```python
# Feedback score -> replcae null value mode
mode = data['Feedback_Score'].mode()[0]
data['Feedback_Score'].fillna(mode, inplace=True)
```

```python
# City ,  country, gender
# replcae all null value by Mode
for col in ['Gender', 'City', 'Country']:
  data[col].fillna(data[col].mode()[0], inplace=True )
```

```python
# Last Purchase Date -> missing value by forward fill

data['Last_Purchase_Date'].ffill(inplace=True )
```

```python
# No more null value
data.isnull().sum()
```

## **50 Correcting Data type**

* Numeric -> int/float
* Categorical -> object

```python
# data types
# Clean and convert the 'Age' column to integer
data['Age'] = data['Age'].astype(str).str.strip() # Ensure all are strings and remove extra spaces
data['Age'] = pd.to_numeric(data['Age'], errors='coerce') # Convert to numeric, coercing errors to NaN
data['Age'].fillna(age_median, inplace=True) # Fill any NaNs (including those from '250', '3', '-10' if problematic) with the median
data['Age'] = data['Age'].astype('int64') # Convert to integer
```

## **6. Handling Outliers**
* Genuine Outliers
* Errors

Remove Outliers by JQR, Z-Score, Domain Knowledge, Transforming data

```python
# plot -> boxplot

import matplotlib.pyplot as plt
import seaborn as sns

cols = ['Age', 'Purchase_Amount', 'Feedback_Score']

plt.figure(figsize=(15, 5))

for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=data[col], color='skyblue', width=0.3)
    plt.title(f'Boxplot of {col}', fontsize=12)
    plt.ylabel('')
    plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
from scipy import stats
import numpy as np
# Calculate Z-scores
z_scores = np.abs(stats.zscore(data[['Age', 'Purchase_Amount', 'Feedback_Score']]))

# Identify outliers (any row with z > 3)
outlier_mask = (z_scores > 3).any(axis=1)
print(f"Number of outliers detected (Z-score method): {outlier_mask.sum()}")

# Display sample outliers
data[outlier_mask]
```

```python
# ~ -> not symbol
data_clean = data[~outlier_mask]
```

```python
cols = ['Age', 'Purchase_Amount', 'Feedback_Score']

plt.figure(figsize=(15, 5))

for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=data_clean[col], color='skyblue', width=0.3)
    plt.title(f'Boxplot of {col}', fontsize=12)
    plt.ylabel('')
    plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# save clean data
data_clean.to_csv('clean_data.csv', index=False)
```
