## Text Data Preprocessing



### 1. **Text Cleaning**

Remove unwanted elements:

* HTML tags → `"<p>hello</p>" → "hello"`
* URLs → `"visit https://example.com" → "visit"`
* Emojis, symbols → `"Good 😊" → "Good"`
* Extra whitespaces, numbers, etc.

*Python (using regex):*

```python
import re
text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # remove URLs
text = re.sub(r'[^A-Za-z\s]', '', text)  # keep only letters
```



### 2. **Lowercasing**

Make all words lowercase to avoid treating "Apple" and "apple" as different.

```python
text = text.lower()
```



### 3. **Tokenization**

Split text into **words or sentences**.

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize("I love NLP!")
# Output: ['I', 'love', 'NLP', '!']
```



###  4. **Stopword Removal**

Remove **common words** like *“the, is, in, a, and”* that add little meaning.

```python
from nltk.corpus import stopwords
tokens = [w for w in tokens if w not in stopwords.words('english')]
```



###  5. **Stemming / Lemmatization**

Reduce words to their **base form**.

  * ####  *Stemming* (rule-based):

`playing → play`, `studies → studi`

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmer.stem('playing')  # Output: 'play'
```

  * ####  *Lemmatization* (dictionary-based):

`studies → study`, `better → good`

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize('studies')  # Output: 'study'
```



### 6. **Text Representation (Feature Extraction)**

Convert text → numbers for ML models.

Common methods:

* **Bag of Words (BoW)**
* **TF-IDF (Term Frequency–Inverse Document Frequency)**
* **Word Embeddings (Word2Vec, GloVe, BERT, GPT)**
