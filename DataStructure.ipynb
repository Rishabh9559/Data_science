**Data Structure**

List - list like a array. Store any data types value <br> List be mutable

```python
li=[]  # empty list
li1=[1,2,"three",4]
print(li1)

li2=["one",[1,2,4],'five',[6,7,8]]   # nested list
print(len(li2))   # length
print(li2[1])  # index 1
print(li2[1][2])  # nested list


## negative index
print(li2[-1])  # last index -1
print(li2[-3:-1])  # start include and end number not include

print(li2[-1:-4])  # empty output, code run forward, util be define reverse.
```

```python
list1=[10,20,30,50,True,False]
print(max(list1))
print(min(list1))  # mini value is false
print(sum(list1))  # sum all element
```

```python
list1[1]=100  # update number
print(list1)

print(list1*3)  ## three time print
```

**List inbuild function**

```python
# append() -> new element at last index
list1.append(60)
print(list1)

# extend() -> new list added at last, extand return none, it only edit list
list1.extend(li1)
print(list1)

# insert() -> two parameter index,value
list1.insert(0,45632)
print(list1)
```

```python
# + Operators -> new list created then add both list into new list
list1=['apple','banana']
list2=['mangoes','cherry']
lis=list1+list2  # add two list and create new list
print(lis)
```

```python
## remove() -> delete first occure first element
list1=['apple','banana','mangoes','cherry','banana']
list1.remove('banana')
print(list1)
```

pop() -> return pop element

```python
print(list1.pop(2))   # give index num to pop element
print(list1)
print(list1.pop())  # bydefault last index taken
print(list1)
```

del keywords -> delete whole list

```python
li1=[1,8,6,'fgj','poj']
del li1[1]
print(li1)
# del li1   # delete list from memory
# print(li1)
```

clear() -> clear element from list not delete list from memory

```python
li1.clear()
print(li1)
```

sorting list

```python
l=[8,1,5,6,1,3]
l.sort()  # sort number
print(l)

l.reverse() # reverse list
print(l)

```

count() -> frequency count

```python
print(l.count(1))
```

joint() -> joint words
