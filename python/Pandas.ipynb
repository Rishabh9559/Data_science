Attributes in serise data structure


```python
data = ['New Delhi', 'Washington', 'London', 'Paris']
index = ['India', 'USA', 'UK', 'France']
series_capital=pd.Series(data,index) # data , index
print(series_capital)
```

```python
#  only index display
print("index only")
print(series_capital.index)
```

```python
#  only value
print("value only")
print(series_capital.values)
```
