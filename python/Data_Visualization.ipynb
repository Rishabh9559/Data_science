```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data_customers=pd.read_excel("/content/customers.xlsx")
data_purchase=pd.read_excel("/content/purchases.xlsx")
data_company=pd.read_excel("/content/products.xlsx")

# Convert relevant columns to string where needed (from previous steps)
data_purchase=data_purchase.astype(str)
data_customers=data_customers.astype(str)


# Clean the 'cost' column in data_company: remove '$' and convert to numeric
# Check if the column exists before attempting to clean it
if 'cost' in data_company.columns:
    # Remove the '$' sign and convert the column to a float
    # Ensure the column is treated as string before replacing
    data_company['cost'] = data_company['cost'].astype(str).str.replace('$', '', regex=False).astype(float)
else:
    print("Warning: 'cost' column not found in data_company.")


# Calculate average cost per company
# This operation will now work because the 'cost' column is numeric
if 'company' in data_company.columns and 'cost' in data_company.columns:
    avg_cost_per_company = data_company.groupby('company')['cost'].mean().reset_index()

    plt.figure(figsize=(18,12))

    plt.bar(avg_cost_per_company['company'], avg_cost_per_company['cost'], color='green')
    plt.title('Average product cost per company', fontsize=16)
    plt.xlabel('Company', fontsize=12)
    plt.ylabel('Average Cost', fontsize=12)

    plt.xticks(rotation=45)
    # Corrected linestyle argument
    plt.grid(axis='y', linestyle='--', linewidth=0.9)
    plt.savefig('average_product_cost_per_company.png')
else:
    print("Warning: 'company' or 'cost' column not found in data_company. Cannot create plot.")
```

```python
plt.figure(figsize=(8,6) )
plt.boxplot(data_company['cost'])

plt.title("Box plot of cost of the product",fontsize=16)
plt.ylabel("cost of the product")
plt.savefig("boxplot_cost_product.png")
plt.show( )
```

**Gender Distribution: which graph is best - pie chart**

```python
gender_count=data_customers[data_customers['gender'].notna()]['gender'].value_counts()
```

```python
print(gender_count)
```

```python
# pie chart of gender
plt.figure(figsize=(4,4))
plt.pie(gender_count,autopct="%1.1f%%",labels=gender_count.index)
plt.title("Gender Distribution")
plt.savefig("gender_distribution.png")
plt.show()
```

**Seaborn**

Seaborn is a Python data visualization library built on top of Matplotlib.

It provides a high-level interface for creating informative and attractive statistical graphics.

```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
# total tips per day
sns.barplot(x='day' , y='tip' , data=tips , estimator=sum)
```

```python
sns.kdeplot(data=tips,x='total_bill', hue='time')
plt.title('KDE plot of the total bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.savefig('kde_total_bill.png')
plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill', hue='time', fill=bool)
plt.title('KDE plot of the total bill')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.savefig('kde_total_bill.png')
plt.show()
```
