Matrix Transpose  : convert row into column and column into row

```python
arr=np.array([[1,5,3],[4,5,6],[6,7,8]])
print(arr)
print('transpose')
print(arr.transpose())
```

```python
def transpose_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    transposed = [[0 for _ in range(rows)] for _ in range(cols)]
    for i in range(rows):
        for j in range(cols):
            transposed[j][i] = matrix[i][j]
    return transposed

matrix = [[1, 2, 3], [4, 5, 6]]
# transposed_matrix = transpose_matrix(matrix)
print(matrix) # Output: [[1, 4], [2, 5], [3, 6]]
```
