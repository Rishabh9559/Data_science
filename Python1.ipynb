```python
a=4
b=6
print(a<b or b==a)   # any condition true return - true.
                      # first conditon check first then second check
print(a<b and a==b)
```

**Bit Operators**

```python
a=4
b=6
print(a|b)  #  | -> OR bits operators
print(a&b)  # & -> And operators
print(a^b)  # ^ XOR operators
print(~a)   # ~ not operators
#  Left shift and Right Shift
print("left shift",a<<b)
print("right shift ",8>>1)
```

len() -> string length <br>
string trim -> varibaleName[start:end] point <br>
String strip name[start:end:skip] <br> Screenshot 2025-04-19 125333.png

```python
print(len(name))  # string length len()
print(name[1:6]) # string split or trime or substring
# index start with 0,1,2,3,.....
print(name[0:8:1]) #by defalut skip be 1 -> 0,1,2,3,4,5,6,7,8
print(name[0:8:2]) # skip by 2 -> 0,2,4,6
print(name[0: ]) # start 0 end last index
print(name[:5]) # by default start 0
```

**Negative string trim strip**

```python
name="hello world where"
print(len(name))
print(name[-13: ])
print(name[:-6])
print(name[ : : -1])  ## reserve string
print(name[: :-2])  ## reverse string  with skip one index
arr=[1,2,3]
print(arr[: : -1])  ## reverse array
```
