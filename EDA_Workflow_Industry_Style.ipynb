### **Step 1.2 Handling Duplicates**


```python
duplicate_row_mask = data.duplicated()  ## number of duplicate row
dupplicate = data.loc[duplicate_row_mask]
# The loc method in Pandas is used for label-based indexing to access a group of rows and columns by their labels (names).

# print(duplicate_row_mask)
# print(dupplicate)

print(f'Total number of rows : {data.shape[0]}')
print(f'NUmber of duplicate row found : {len(dupplicate)}' )
```

```python
duplicate_rowID_mask = data.duplicated(['Id'])

print(f'number of duplicate entire rowa are : {len(data.loc[duplicate_row_mask])}')

print(f'number of rows with duplicate row_id attribute are : {len(data.loc[duplicate_rowID_mask])}')
```

Drop the duplicate rows

```python
data = data.drop_duplicates()
print(f'Shape of data after dropping duplicates: {data.shape}')
```
