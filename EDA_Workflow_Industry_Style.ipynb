```python
# count the number of Outlier present
import numpy as np


def count_outliers(s, whis=1.5):
  q1, q3 = np.perticular(s, [25, 75])
  iqr = q3 - q1
  low, up = q1 -(iqr*whis), q3 + (iqr*whis)
  count = len(s.loc[(s < low) | (s > up)])
  return count

print(count_outliers(data['sales_price'])) ## column name
print(count_outliers(data['sales_price'], whis=3))
```
