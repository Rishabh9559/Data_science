# **Logistic Regression**

**Logistic Regression** is a **supervised machine learning algorithm** used for **classification tasks** â€” most commonly **binary classification** (i.e., classifying data into two categories, such as spam/not spam, disease/no disease, etc.).



##### **Key Idea**

Unlike **Linear Regression**, which predicts continuous values, **Logistic Regression predicts the probability** that a given input belongs to a particular class.



#### **How It Works**

1. **Linear Combination**
   It starts with a linear equation:

   $$
   z = w_1x_1 + w_2x_2 + \dots + w_nx_n + b
   $$

   where:

   * $x_i$: input features
   * $w_i$: weights
   * $b$: bias

2. **Sigmoid Function (Logistic Function)**
   It then applies a **sigmoid function** to map this output $z$ into a range between 0 and 1:

   $$
   \sigma(z) = \frac{1}{1 + e^{-z}}
   $$

   This gives a **probability** that the input belongs to class 1.

3. **Threshold**
   If the probability $\geq 0.5$, classify as class 1; else, class 0 (this threshold can be adjusted).



####  **Use Cases**

* Email spam detection
* Disease diagnosis (e.g., cancer: yes/no)
* Credit card fraud detection
* Customer churn prediction



####  **Loss Function**

Logistic regression uses **Binary Cross-Entropy (Log Loss)**:

$$
\text{Loss} = -\left[ y \cdot \log(\hat{y}) + (1 - y) \cdot \log(1 - \hat{y}) \right]
$$

Where:

* $y$ is the true label (0 or 1)
* $\hat{y}$ is the predicted probability



####  **Advantages**

* Simple and fast
* Interpretable (you can examine feature weights)
* Works well with linearly separable data



####  **Limitations**

* Can't handle non-linear relationships directly
* May underperform if feature interactions are complex
