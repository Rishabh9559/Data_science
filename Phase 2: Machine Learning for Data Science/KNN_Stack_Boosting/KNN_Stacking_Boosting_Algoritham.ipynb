# KNN (K-Nearest Neighbors)

image.png

KNN, or K-Nearest Neighbors, is a supervised machine learning algorithm used for both classification and regression tasks. Its core idea is that data points with similar features tend to be located near one another, so the algorithm predicts the label or value for a new data point based on the labels or values of its closest neighbors in the training data.

## How KNN Works
* Select k (number of neighbors): You choose the number of closest points (neighbors) to consider. This is a key parameter in the algorithm.

* Calculate distance: For a given test data point, KNN calculates the distance to all points in the training dataset using a distance metric like Euclidean or Manhattan distance.

* Choose neighbors: It finds the k closest data points based on the calculated distances.

* Make prediction:

  * Classification: Assigns the new point to the category most common among its k nearest neighbors (majority/plurality vote).

  * Regression: Predicts the value by averaging the values of the k nearest neighbors
