**KNN**, or K-Nearest Neighbors, is a **supervised** machine learning algorithm used for **both classification and regression tasks**. Its core idea is that data points with **similar features** tend to be located near one another, so the algorithm predicts the label or value for a new data point based on the labels or values of its closest neighbors in the training data.


### K -> number of neighbors

## How KNN Works
* Select k (number of neighbors): You choose the number of closest points (neighbors) to consider. This is a key parameter in the algorithm.

* Calculate distance: For a given test data point, KNN calculates the distance to all points in the training dataset using a distance metric like Euclidean or Manhattan distance.

* Choose neighbors: It finds the k closest data points based on the calculated distances.

* Make prediction:

  * Classification: Assigns the new point to the category most common among its k nearest neighbors (majority/plurality vote).

  * Regression: Predicts the value by averaging the values of the k nearest neighbors


##  What is KNN?

* **KNN (K-Nearest Neighbors)** is a **supervised machine learning algorithm** used for **classification** and sometimes **regression**.
* It is a **lazy learner algorithm**  it doesnt explicitly train a model, instead it stores the training data and makes predictions based on similarity (distance).
* It is an **instance-based learning algorithm**  prediction is based on the nearest data points.



##  Working Principle

1. Choose a value of **K** (number of neighbors).
2. Compute the **distance** (e.g., Euclidean distance) between the test sample and all training samples.
3. Select the **K nearest neighbors**.
4. Perform:

   * **Classification**  majority voting of neighbors class.
   * **Regression**  average (or weighted average) of neighbors values.


##  Formula

The most common **distance formula** used is **Euclidean Distance**:

$$
d(p, q) = \sqrt{\sum_{i=1}^n (p_i - q_i)^2}
$$

Where:

* $p$ and $q$ are two points in n-dimensional space.
* $p_i, q_i$ are feature values.

Other distance measures:

* **Manhattan Distance**: $|p_i - q_i|$
* **Minkowski Distance**: generalized version.
* **Cosine Similarity** (for text and vectors).

##  Example

Suppose we want to classify a fruit as **Apple** or **Orange** based on:

* Weight
* Color

| Fruit | Weight (g) | Color Score (1=Green, 2=Yellow, 3=Red, 4=Orange) | Label  |
| ----- | ---------- | ------------------------------------------------ | ------ |
| A     | 150        | 3                                                | Apple  |
| B     | 170        | 3                                                | Apple  |
| C     | 140        | 4                                                | Orange |
| D     | 130        | 4                                                | Orange |

 Test Point: (160, 3)

* Compute distances with all points.
* If **K = 3**, then nearest 3 neighbors are \[A, B, C].
* Majority class = **Apple**.
   Prediction = Apple.



##  Types of KNN

1. **KNN Classification**  predicts class labels (e.g., Spam / Not Spam).
2. **KNN Regression**  predicts continuous values (e.g., house prices).
3. **Weighted KNN**  closer neighbors have more weight.
4. **Distance-based KNN**  different distance metrics (Euclidean, Manhattan, Cosine).


##  Features of KNN

* Simple and intuitive.
* Non-parametric (no assumptions about data distribution).
* Works well with small datasets.
* Sensitive to:

  * Value of **K**.
  * Distance metric used.
  * Irrelevant features (so **feature scaling** is important).

## Ball tree and KD Tree Algorithms

1. KD- tree:

A hierarchical data structure called a KD-Tree (K-Dimensional Tree) is used for multidimensional space partitioning. It creates a binary tree with each node representing a part of the space by iteratively dividing the space along predefined dimensions. Often used for effective closest neighbor searches.

image.png

2. Ball tree:

It is a tree-based data structure designed for organizing and querying points in multidimensional spaces. It is also a binary tree with a hierarchical (binary) structure. The ball tree data structure is particularly effective when there are a lot of dimensions. Unlike KD-trees, Ball trees use hyperspheres to represent partitions, grouping nearby points within each hypersphere.
