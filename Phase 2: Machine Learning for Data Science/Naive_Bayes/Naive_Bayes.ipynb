# Naive Bayes

image.png

Naive Bayes is a **probabilistic machine learning algorithm** based on **Bayes Theorem**, commonly used for classification tasks.

###  Key Idea:

It assumes that all features are **independent** of each other given the class label  which is usually not true in real-world data, but the assumption ("naive") works surprisingly well in practice.


###  Bayes Theorem:

$$
P(C \mid X) = \frac{P(X \mid C) \cdot P(C)}{P(X)}
$$

Where:

* $P(C \mid X)$: Posterior probability of class $C$ given input features $X$.
* $P(X \mid C)$: Likelihood (probability of input features given class $C$).
* $P(C)$: Prior probability of class $C$.
* $P(X)$: Evidence (probability of the input features).

Naive Bayes simplifies this by assuming features are conditionally independent:

$$
P(X \mid C) = P(x_1 \mid C) \cdot P(x_2 \mid C) \cdot \dots \cdot P(x_n \mid C)
$$



###  Types of Naive Bayes:

1. **Gaussian Naive Bayes**  assumes features follow a normal (Gaussian) distribution.
2. **Multinomial Naive Bayes**  good for text classification (e.g., spam detection).
3. **Bernoulli Naive Bayes**  works with binary features (e.g., word present or not).



###  Advantages:

* Very fast to train and predict.
* Works well with high-dimensional data (like text classification).
* Simple and easy to implement.

###  Limitations:

* Assumes independence among features (rarely true).
* Struggles when features are correlated.
* Can give poor probability estimates, though classification is often good.


 **Example Use Case**:

* **Spam Detection**:

  * Features = words in an email.
  * Class = Spam or Not Spam.
  * Naive Bayes calculates probabilities of spam vs. non-spam and picks the higher one.
