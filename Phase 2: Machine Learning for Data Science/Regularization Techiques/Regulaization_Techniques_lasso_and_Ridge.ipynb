### Regularization Techniques: Lasso and Ridge Regression

**Regularization** is a technique used in machine learning to prevent **overfitting** by adding a penalty term to the loss function. The goal is to keep the model weights small and avoid relying too much on any one feature.

Two of the most common regularization techniques in linear regression are:

---

## üîπ 1. **Ridge Regression (L2 Regularization)**

* **Adds penalty:**

  $$
  \text{Loss Function} = \text{MSE} + \lambda \sum_{j=1}^{n} w_j^2
  $$

  where:

  * $\text{MSE}$ = Mean Squared Error
  * $w_j$ = model coefficients
  * $\lambda$ = regularization strength (hyperparameter)

* **Effect:** It **shrinks coefficients** towards zero but **does not make them exactly zero**. Useful when many small/medium effects exist.

* **Use case:** Works well when all features are relevant or collinear (e.g., multicollinearity problems).

---

## üîπ 2. **Lasso Regression (L1 Regularization)**

* **Adds penalty:**

  $$
  \text{Loss Function} = \text{MSE} + \lambda \sum_{j=1}^{n} |w_j|
  $$

* **Effect:** It **shrinks some coefficients to exactly zero**, performing **feature selection**.

* **Use case:** Useful when you suspect only a **few features are important** (sparse solutions).

---

## üîÑ Comparison:

| Feature               | Ridge (L2)           | Lasso (L1)                             |      |   |
| --------------------- | -------------------- | -------------------------------------- | ---- | - |
| Penalty Term          | $\lambda \sum w_j^2$ | ( \lambda \sum                         | w\_j | ) |
| Feature Selection     | ‚ùå No                 | ‚úÖ Yes                                  |      |   |
| Coefficients become 0 | ‚ùå No                 | ‚úÖ Yes (some)                           |      |   |
| Best for              | Many small effects   | Few important features                 |      |   |
| Stability             | More stable          | Less stable if features are correlated |      |   |

---

## üîÇ Elastic Net: Hybrid of Both

* Combines L1 and L2:

  $$
  \text{Loss} = \text{MSE} + \alpha \left( \lambda_1 \sum |w_j| + \lambda_2 \sum w_j^2 \right)
  $$
* Useful when you want both **shrinkage** and **feature selection**.

---

##

* **Lasso**: Great for reducing dimensionality by selecting important features.
* **Ridge**: Great when all features are useful but may be collinear.
* **Elastic Net**: When you want a balance between Lasso and Ridge.

## üìâ 1. **What is Underfitting**

* **Definition**:
  The model is **too simple** to capture the underlying patterns in the data.

* **Symptoms**:

  * High error on **training** data
  * High error on **testing** data
  * Poor performance everywhere

* **Causes**:

  * Model is too simple (e.g., linear model on complex data)
  * Not enough features
  * Too much regularization
  * Not enough training time

* **Example**:
  Predicting house prices using only the number of rooms, ignoring location, size, etc.

---

## üìà 2. **What is Overfitting**

* **Definition**:
  The model is **too complex** and fits the **noise** in the training data instead of generalizing.

* **Symptoms**:

  * Very low error on **training** data
  * High error on **testing** data

* **Causes**:

  * Too many features or parameters
  * Too few training examples
  * Lack of regularization
  * Training too long

* **Example**:
  A model that memorizes the training dataset but performs poorly on new, unseen data.

---

## üîÑ Comparing Overfitting vs Underfitting

| Feature           | Underfitting        | Overfitting        |
| ----------------- | ------------------- | ------------------ |
| Model Complexity  | Too low (simple)    | Too high (complex) |
| Training Accuracy | Low                 | Very high          |
| Test Accuracy     | Low                 | Low                |
| Generalization    | Poor                | Poor               |
| Cause             | Too little learning | Too much learning  |

---

## üìä Ideal: **Good Fit**

* **Balanced model** that captures the pattern in the data but doesn‚Äôt memorize it.
* **Low training error** and **low validation/test error**

---

## üõ†Ô∏è How to Fix

* ‚úÖ **Underfitting**:

  * Use more complex model
  * Add more features
  * Reduce regularization
  * Train longer

* ‚úÖ **Overfitting**:

  * Use simpler model
  * Use regularization (L1/L2)
  * Use more training data
  * Apply dropout (for neural networks)
  * Use early stopping

---






##  What is **Noise** in Machine Learning?

**Noise** refers to **irrelevant or random variations** in data that do **not represent true underlying patterns**.

---

### üîç Examples of Noise:

* Measurement errors in sensors
* Typing mistakes in data entry
* Irrelevant data (e.g., including a user‚Äôs IP address in a movie recommendation model)
* Outliers that don‚Äôt follow the general pattern

---

### ‚ö†Ô∏è Why Noise is a Problem:

* Makes the model learn **incorrect patterns**
* Can lead to **overfitting**, especially if the model is too complex
* Reduces **accuracy** on unseen (test) data

---

### üßπ How to Reduce Noise:

* Data cleaning (removing outliers, correcting errors)
* Feature selection
* Smoothing techniques or filters
* Regularization techniques (like Lasso, Ridge)

---

## üß† What is **Generalizing**?

In machine learning, **generalization** refers to a model's ability to accurately predict outcomes on new, unseen data after being trained on a specific dataset. A model that generalizes well can apply the patterns it learned from the training data to make reliable predictions on data it has not encountered before

---

### üîç Why It Matters:

* A model that generalizes well will work on **real-world data**, not just memorized training samples.
* Generalization is the **goal** of most ML models ‚Äî we want them to **learn patterns**, not just remember the training set.

---

### üéØ Good Generalization:

* Low error on **training** data ‚úÖ
* Low error on **test/validation** data ‚úÖ
* Model learned the **true relationship** between input and output ‚úÖ

---

### üîÅ Bad Generalization:

* Happens in **overfitting** ‚Äî the model memorizes training noise instead of learning real patterns.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.DataFrame(housing.data, columns=housing.feature_names)
dataset
```

```python
dataset['Price'] = housing.target
dataset.head()
```

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
y
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
```
