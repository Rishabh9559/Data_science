##  **1. What is Linear Regression?**

**Linear Regression** is a **supervised learning** algorithm used for **predicting a continuous value**.

It models the relationship between:

* üìà **Independent variable(s)** (input `x`)
* üìâ **Dependent variable** (output `y`)

### üîß Equation of a Line:

$$
y = mx + c \quad \text{or} \quad y = w_1x + w_0
$$

Where:

* `y` = predicted value
* `x` = input feature
* `w‚ÇÅ` = weight (slope)
* `w‚ÇÄ` = bias/intercept

---

##  **2. Goal of Linear Regression:**

Find the **best fit line** (values of `w‚ÇÅ` and `w‚ÇÄ`) that **minimizes the error** between predicted and actual values.

---

##  **3. What is Gradient Descent?**

**Gradient Descent** is an **optimization algorithm** used to minimize a **loss function** (like MSE - Mean Squared Error).

###  Loss Function (MSE):

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

Where:

* $y_i$ = actual value
* $\hat{y}_i$ = predicted value
* `n` = number of data points

---

##  **4. How Gradient Descent Works:**

1. Start with random weights (`w‚ÇÄ`, `w‚ÇÅ`)
2. Calculate **error/loss** using MSE
3. Compute **gradients** (derivatives of loss w\.r.t weights)
4. **Update weights**:

   $$
   w := w - \alpha \cdot \frac{\partial \text{Loss}}{\partial w}
   $$

   where `Œ±` is the **learning rate**
5. Repeat until convergence (loss is minimal)

---

##  Example (1D Linear Regression):

Given data:

```
x: [1, 2, 3, 4]
y: [2, 4, 6, 8]
```

Initialize `w‚ÇÄ=0`, `w‚ÇÅ=0`

* Predict $\hat{y}_i = w_1x_i + w_0$
* Use gradient descent to update `w‚ÇÅ` and `w‚ÇÄ` to minimize error

Eventually, it will learn:

$$
\hat{y} = 2x \quad (\text{ideal line})
$$

## **MSE - Mean Square Erro**


MSE is a common **loss function** used in **regression problems** to measure the **average of the squared differences** between actual and predicted values.

---

###  **MSE Formula:**

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

Where:

* $y_i$ = Actual value
* $\hat{y}_i$ = Predicted value
* $n$ = Total number of data points

---

### **Why Square the Error?**

* Avoids negative errors cancelling out positive ones.
* Penalizes large errors more heavily than small ones.

---

###  Example:

Let‚Äôs say:

* Actual values $y = [3, 5, 7]$
* Predicted values $\hat{y} = [2, 5, 10]$

$$
\text{MSE} = \frac{1}{3}[(3-2)^2 + (5-5)^2 + (7-10)^2] \\
= \frac{1}{3}[1 + 0 + 9] = \frac{10}{3} \approx 3.33
$$

## üìè 1. **Best Fit Line** (in Linear Regression)

The **best fit line** is the line that **best represents the data** in a **scatter plot** by minimizing the **distance between the actual points and the predicted line**.

### üîß Equation of the best fit line:

$$
y = w_1x + w_0
$$

Where:

* $x$: input (independent variable)
* $y$: output (dependent variable)
* $w_1$: slope
* $w_0$: intercept

### üéØ Goal:

Minimize the **total error** between predicted values ($\hat{y}$) and actual values ($y$).

We use algorithms like **Gradient Descent** to adjust `w‚ÇÅ` and `w‚ÇÄ` so that this line fits the data as accurately as possible.

---

##  2. **Residual Error**



### üîç What is Residual Error?

Residual Error is the **difference between the actual value** and the **predicted value** from the model.

###  Formula:

$$
\text{Residual} = y_i - \hat{y}_i
$$

Where:

* $y_i$: actual value
* $\hat{y}_i$: predicted value from the best fit line

Each data point has its **own residual**.

---

### üîé Visual Intuition:

* If a point lies **above** the line ‚Üí residual is **positive**
* If it lies **below** the line ‚Üí residual is **negative**
* If it lies **on** the line ‚Üí residual is **zero**

##  **1. Convex Curve**

A **convex curve** is a type of curve where:

> **Any line segment between two points on the curve lies above or on the curve itself.**

###  Mathematically:

A function $f(x)$ is **convex** if:

$$
f(\lambda x_1 + (1 - \lambda) x_2) \leq \lambda f(x_1) + (1 - \lambda) f(x_2)
$$

for all $x_1, x_2$ and $\lambda \in [0,1]$

---



### ‚úÖ **Global Minimum**:

* The **lowest point** of the entire function.
* No other point has a smaller function value.

### ‚úÖ **Local Minimum**:

* The **lowest point** in a small neighborhood.
* May not be the lowest overall.

---


## 2. For Non-Convex Functions:

* The function may have **multiple local minima**
* Only one of them is the **global minimum**
* Gradient Descent may get **stuck in a local minimum**

---



| Term            | Meaning                               |
| --------------- | ------------------------------------- |
| Convex Function | Curve where every local min is global |
| Local Minimum   | Lowest point in a small neighborhood  |
| Global Minimum  | Lowest point in the entire function   |
| In Convex Curve | Local Minimum = Global Minimum        |

## üìò **Loss Function vs. Cost Function**

| Aspect                   | **Loss Function**                                      | **Cost Function**                                      |
| ------------------------ | ------------------------------------------------------ | ------------------------------------------------------ |
| **Definition**           | Measures error for **a single training example**       | Measures **average error** over the **entire dataset** |
| **Use**                  | Used during backpropagation (one data point at a time) | Used to evaluate model performance overall             |
| **Scope**                | Instance-level                                         | Dataset-level                                          |
| **Example (Regression)** | $L(y_i, \hat{y}_i) = (y_i - \hat{y}_i)^2$              | $J(w) = \frac{1}{n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2$ |

---

## üîç Explanation with Example:

### üéØ Suppose you're predicting house prices:

### üì¶ Dataset:

| Input (sqft) | Actual Price (Y) | Predicted (≈∑) |
| ------------ | ---------------- | ------------- |
| 1000         | ‚Çπ50 Lakh         | ‚Çπ48 Lakh      |
| 1200         | ‚Çπ60 Lakh         | ‚Çπ62 Lakh      |

---

### ‚úÖ **Loss Function** (for one sample):

$$
\text{Loss} = (Y - \hat{Y})^2
$$

For 1st house:

$$
(50 - 48)^2 = 4
$$

---

### ‚úÖ **Cost Function** (average over all):

$$
\text{Cost} = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2
$$

For both houses:

$$
\text{Cost} = \frac{(50 - 48)^2 + (60 - 62)^2}{2} = \frac{4 + 4}{2} = 4
$$

---

## üß† Summary:

| Term         | Loss Function            | Cost Function                     |
| ------------ | ------------------------ | --------------------------------- |
| Data Size    | Single example           | Entire dataset                    |
| Purpose      | Measure per-sample error | Measure model performance overall |
| Optimization | Used to compute gradient | Guides learning process           |
| Output       | Scalar value             | Scalar value (average loss)       |

---


In practice, especially in deep learning:

* "Loss function" is often used to refer to **both** loss and cost interchangeably.
* But technically, the **cost function = average of individual losses**.

## üìò **Key Assumptions of Linear Regression:**

|  # | Assumption               | Description                                                                                                                     |
| -: | ------------------------ | ------------------------------------------------------------------------------------------------------------------------------- |
|  1 | **Linearity**            | The relationship between the independent variable(s) (X) and the dependent variable (Y) is **linear**.                          |
|  2 | **Independence**         | Observations (data points) are **independent** of each other. No autocorrelation.                                               |
|  3 | **Homoscedasticity**     | The **residuals (errors)** have **constant variance** across all levels of the independent variable.                            |
|  4 | **Normality of Errors**  | The residuals (not the variables) are **normally distributed**. Important for inference (e.g., p-values, confidence intervals). |
|  5 | **No Multicollinearity** | Independent variables should **not be highly correlated** with each other.                                                      |
|  6 | **No Endogeneity**       | Independent variables should **not be correlated with the error term**.                                                         |

---

### üîç Explained with Simple Terms:

---

### 1. üìà **Linearity**

‚úÖ Assumes a straight-line relationship between X and Y.
**Check**: Use scatter plots.

---

### 2. üîó **Independence of Observations**

‚úÖ Data points must not influence each other.
**Check**: Use **Durbin-Watson Test** for time series data.

---

### 3. üîÑ **Homoscedasticity**

‚úÖ Spread (variance) of residuals should be consistent across all values of X.
**Check**: Plot residuals vs. predicted values.

---

### 4. üìä **Normality of Residuals**

‚úÖ Errors should follow a normal distribution.
**Check**: Histogram or Q-Q plot of residuals.

---

### 5. üîÅ **No Multicollinearity**

‚úÖ Input variables (X1, X2, ...) should not be too similar.
**Check**: Calculate **VIF (Variance Inflation Factor)**. VIF > 10 indicates multicollinearity.

---

### 6. üö´ **No Endogeneity**

‚úÖ Independent variables must not be correlated with the error term (no hidden bias).
**Violation leads to biased coefficients.**

---

## üìò **Definition:**

**Polynomial Regression** is an extension of **Linear Regression** where the relationship between the independent variable `x` and the dependent variable `y` is modeled as an **nth-degree polynomial**.

> ‚úÖ It allows for **curved relationships** between input and output, unlike simple linear regression which fits a straight line.

---

### üßÆ **Polynomial Equation:**

$$
y = w_0 + w_1x + w_2x^2 + w_3x^3 + \cdots + w_nx^n
$$

Where:

* $x$: Input feature
* $y$: Target/output
* $w_0, w_1, ..., w_n$: Coefficients learned during training
* $n$: Degree of the polynomial (2 = quadratic, 3 = cubic, etc.)

---

## üìà **When to Use Polynomial Regression?**

* When the data shows a **non-linear relationship**.
* If linear regression underfits your data (i.e., poor performance on both training and test sets).

---

## üîÅ **Linear vs. Polynomial Regression:**

| Feature                | Linear Regression | Polynomial Regression           |
| ---------------------- | ----------------- | ------------------------------- |
| Equation Type          | $y = w_0 + w_1x$  | $y = w_0 + w_1x + w_2x^2 + ...$ |
| Shape of Curve         | Straight line     | Curved (parabola, cubic, etc.)  |
| Captures Non-Linearity | ‚ùå No              | ‚úÖ Yes                           |

---


## ‚ö†Ô∏è **Overfitting Risk**

* Higher-degree polynomials can fit training data **too well** and perform **poorly on new data**.
* **Use cross-validation** to choose the best degree.

---

## ‚úÖ Summary

| Term                  | Meaning                                           |
| --------------------- | ------------------------------------------------- |
| Polynomial Regression | Models non-linear relationships using polynomials |
| Degree                | Power of the highest term (e.g., x¬≤, x¬≥, etc.)    |
| Advantage             | Captures curves in data                           |
| Risk                  | Can overfit if degree is too high                 |
