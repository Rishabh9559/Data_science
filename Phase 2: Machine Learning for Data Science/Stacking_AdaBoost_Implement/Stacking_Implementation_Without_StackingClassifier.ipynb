Stacking_Implementation_Without_StackingClassifier

```python
# Import libery
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
data = pd.read_csv("/content/diabetes.csv")
```

```python
data
```

```python
# data is imbalance
print(data['Outcome'].value_counts())
```

```python
# split the data input feature and target feature
X = data.drop(columns=['Outcome'],axis=1)
y = data['Outcome']
```

Split the entire data into two subset

* Subset 1 Train set
* Subset 2 Validation set

BY change the value of Random State the model accuracy may incress

```python
from sklearn.model_selection import train_test_split

train, validate_train, test, validate_test = train_test_split(X, y, test_size=0.5, random_state=355)
```

```python
# split the train and test
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.2, random_state=355)
```

Model 1: Implementaion of KNeighboursClassifier

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
knn_model.score(X_test, y_test)
```

Model 2: Implement of SVC Classifier

```python
svc_model= SVC()
svc_model.fit(X_train, y_train)
svc_model.score(X_test, y_test)
```

Prediction of the model in the validation set

```python
predict_validation_knn = knn_model.predict(validate_train)

predict_validation_svc = svc_model.predict(validate_train)
```

```python
predict_validation = np.column_stack( (predict_validation_knn, predict_validation_svc))

predict_validation
```

```python
predict_test_knn = knn_model.predict(X_test)
predict_test_svc = svc_model.predict(X_test)

predict_test = np.column_stack( (predict_test_knn, predict_test_svc))

predict_test
```

the predict validation adn validatio test give as data in Random forest classifier

Stacking: Input features are the predictions of the model M1(KNN) and model M2(SVC) stored inside the predict_val(input features) for the RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

random_forst_classifier_model = RandomForestClassifier()
random_forst_classifier_model.fit(predict_validation, validate_test)
```

```python
# ckec the score
random_forst_classifier_model.score(predict_test, y_test)
```

Hyper-parameter tuning GridSearchCV

```python
grid_param = {
    "n_estimators" : [90,100,115],
    'criterion': ['gini', 'entropy'],
    'min_samples_leaf' : [1,2,3,4,5],
    'min_samples_split': [4,5,6,7,8],
    'max_features' : ['auto','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search_model = GridSearchCV(estimator=random_forst_classifier_model,param_grid=grid_param,cv=5,n_jobs =-1,verbose = 3)
```

```python
grid_search_model.fit(predict_validation, validate_test)
```

```python
grid_search_model.best_params_
```

```python
random_forst_classifier_hyperparameter_tuning = RandomForestClassifier( n_estimators=90, min_samples_split=4)

random_forst_classifier_hyperparameter_tuning.fit(predict_validation, validate_test)
```

```python
# score
random_forst_classifier_hyperparameter_tuning.score(predict_test, y_test)
```
