# **Retrieving Data with Select Statements**

| Key Name           | Type              | Definition                                                                 |
|--------------------|-------------------|-----------------------------------------------------------------------------|
| PRIMARY KEY        | Uniqueness Key    | Uniquely identifies each row in a table; cannot contain NULL.              |
| FOREIGN KEY        | Relationship Key  | Links two tables; references PRIMARY KEY of another table.                 |
| CANDIDATE KEY      | Uniqueness Key    | Columns that qualify to become a PRIMARY KEY.                              |
| SUPER KEY          | Uniqueness Key    | Any set of columns that uniquely identify a row (including extra columns). |
| COMPOSITE KEY      | Uniqueness Key    | A primary/unique key made of 2 or more columns.                            |
| UNIQUE KEY         | Uniqueness Key    | Ensures values in a column are unique, but allows one NULL.                |
| ALTERNATE KEY      | Uniqueness Key    | Candidate keys that are not selected as PRIMARY KEY.                       |
| CHECK CONSTRAINT   | Validation Key    | Ensures data meets specific conditions.                                    |
| NOT NULL           | Validation Key    | Ensures a column cannot store NULL values.                                 |
| DEFAULT            | Validation Key    | Sets a default value for a column when no value is provided.               |
| INDEX              | Performance Key   | Improves search and query performance on columns.                          |
| FULLTEXT INDEX     | Search Key        | Used for full-text search in large text fields.                            |
| UNIQUE INDEX       | Uniqueness Key    | Ensures indexed column values are unique.                                  |
