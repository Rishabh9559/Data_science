# **CRUD Operations**

C - CREATE

R - READ

U - UPDATE

D - DELETE

### **UPDATE** : UPDATE , SET

Used to update row details from table

```
UPDATE table-name SET column-name = "row-details"  WHERE id=6
```

### **DELETE**
delete key used to delete row from table.

```
DELETE FROM employe-table WHERE id=5;
```

#### ADD

Add new column in Table
```
ALTER TABLE employe-table ADD COLUMN column-name datatype;
```

### **COUNT**

count all row from Tabel
```
SELECT COUNT(*) FROM employe
```

### **AS** Aliases
AS keyword is used to assign a temporary name, or alias

```
SELECT COUNT(*) AS total_employ FROM employe-table;
```

### **MAX**
max value from column

```
SELECT MAX(Salary) AS max-salary FROM employe-table
```

### **MIN**
min value from column
```
SELECT MIN(Salary) AS min-salary
FROM employe-table;
```

### 1. DDL (Data Definition Language)

These commands define or modify the structure of database objects (tables, schemas, indexes, etc.).

* `CREATE TABLE table_name (column1 data_type, column2 data_type, );`  e.g.,

  ````sql
  CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    hire_date DATE
  );
  ``` :contentReference[oaicite:3]{index=3}  
  ````
* `DROP TABLE table_name;`  delete a table.
* `ALTER TABLE table_name ADD COLUMN column_name data_type;`  modify a table structure.
* `TRUNCATE TABLE table_name;`  remove all records (and often reset storage) from a table.
* `COMMENT ON TABLE table_name IS 'comment_text';`  add comment/meta-data.
* `RENAME TABLE old_table_name TO new_table_name;`  rename a table.



### 2. DQL (Data Query Language)

This is primarily about retrieving/querying data from the database.

* `SELECT column1, column2,  FROM table_name WHERE condition;`  basic select.
* Clauses used with SELECT:

  * `WHERE condition`  filter rows.
  * `GROUP BY column1`  group rows by value.
  * `HAVING condition`  filter groups after `GROUP BY`.
  * `DISTINCT`  remove duplicate rows.
  * `ORDER BY column1 [ASC|DESC]`  sort result set.
  * `LIMIT number`  restrict number of rows returned.
* Example:

  ````sql
  SELECT first_name, last_name, hire_date
  FROM employees
  WHERE department = 'Sales'
  ORDER BY hire_date DESC;
  ``` :contentReference[oaicite:17]{index=17}  
  ````


### 3. DML (Data Manipulation Language)

Commands to manipulate the data (insert, update, delete).

* `INSERT INTO table_name (column1, column2, ) VALUES (value1, value2, );`  add a record.
  Example:

  ````sql
  INSERT INTO employees (first_name, last_name, department)
  VALUES ('Jane', 'Smith', 'HR');
  ``` :contentReference[oaicite:20]{index=20}  
  ````
* `UPDATE table_name SET column1 = value1, column2 = value2 WHERE condition;`  modify existing data.
* `DELETE FROM table_name WHERE condition;`  remove records.



### 4. DCL (Data Control Language)

Commands for permissions, access control.

* `GRANT privilege_type [(column_list)] ON [object_type] object_name TO user [WITH GRANT OPTION];`  grant permissions.
  Example:

  ````sql
  GRANT SELECT, UPDATE ON employees TO user_name;
  ``` :contentReference[oaicite:25]{index=25}  
  ````
* `REVOKE [GRANT OPTION FOR] privilege_type [(column_list)] ON [object_type] object_name FROM user [CASCADE];`  revoke permissions.



### 5. TCL (Transaction Control Language)

Commands to manage transactions (commit, rollback, savepoints).

* `BEGIN TRANSACTION;`  start a transaction.
* `COMMIT;`  make changes permanent.
* `ROLLBACK;`  undo changes of transaction.
* `SAVEPOINT savepoint_name;`  set a savepoint inside a transaction.
* Example:

  ````sql
  BEGIN TRANSACTION;
    UPDATE employees SET department = 'Marketing' WHERE department = 'Sales';
    SAVEPOINT before_update;
    UPDATE employees SET department = 'IT' WHERE department = 'HR';
    ROLLBACK TO SAVEPOINT before_update;
  COMMIT;
  ``` :contentReference[oaicite:32]{index=32}  
  ````
