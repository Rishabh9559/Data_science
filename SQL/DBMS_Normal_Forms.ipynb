# **DBMS Normal Forms**

## **Key in Relation**

* **Super key** :
  * The set of one or more attribute (columns) that can uniquely identify a tuple (record or rows) is know as Super key.
  * It support NULL value in rows



* **Candidate Key** :
  * The minimal set of attribute (column) that can uniquely identify a tuple is know as a Candidate key.
  * No repeated data, Unique value
  * It can be NULL value


* **Primary Key** :
  * There can be more than one Candidate key in relation out of which one can chose as Primary Key.
  * One Table contains only One Primary key
  * It can not be NULL value in record/tuple/rows
  * A primary key is Unique key.
  

* **Alternate Key** :
  * An Alternate key is any Candidate key in a tuble/Rows/Records that is not chose as the primary key.
  * It can uniquely identify records
  * It can be consist of one or more attribute/columns.


* **Foreign Key** :
  * A Foreign Key is an attribute in one Table that refers to Primary Key in another Table.
  * The Table that contains Foreign Key is called Referance Table
  * It used to establishing a Relationship between Two Table.

## **1 Normal Form**

  * In 1 normal form each attribule have only one unit of context
  * Each field contains atomic (single) values
  * No repeating groups or multi-valued attrib

  example:
  | StudentID | StudentName | Subjects     |
| --------- | ----------- | ------------ |
| 101       | Rishabh     | DBMS, OS, CN |
| 102       | Amit        | OS, CN       |
| 103       | Neha        | DBMS         |


* Here in subject attribute/ column have more than one value in record so its not 1NF, we have to convert into single pr unit forms.

1NF solution: split table into Base table and Referance table

Base Table:
| StudentID (PK) | StudentName |
| -------------- | ----------- |
| 101            | Rishabh     |
| 102            | Amit        |
| 103            | Neha        |


Referance Table :
| StudentID (FK) | CourseName |
| -------------- | ---------- |
| 101            | DBMS       |
| 101            | OS         |
| 101            | CN         |
| 102            | OS         |
| 102            | CN         |
| 103            | DBMS       |








## **2 Normal Form**

  * Table or relation must be in 1 Normal Form.
  * All the non-prime attribute should be fully-functional dependent on Candidant key.

    <font color="red">**Candidant Key** : A candidate key in a database is a minimal set of one or more attributes (columns) that can uniquely identify each record (row) in a table </font>

  * There should be No Partial dependency in the relation.

Example :
| StudentID | CourseID | StudentName | CourseName | Marks |
| --------- | -------- | ----------- | ---------- | ----- |
| 101       | C1       | Rishabh     | DBMS       | 85    |
| 101       | C2       | Rishabh     | OS         | 90    |
| 102       | C2       | Amit        | OS         | 88    |
| 103       | C1       | Neha        | DBMS       | 92    |

<br>

Candidant key : StudentID + CourseID

<br>

**Functional Dependencies**:

  StudentID → StudentName

  CourseID → CourseName

  (StudentID, CourseID) → Marks

  So:

  StudentName depends only on StudentID

  CourseName depends only on CourseID

  These are partial dependencies → ❌ Not in 2NF

**Base Table** Student Table
| StudentID (PK) | StudentName |
| -------------- | ----------- |
| 101            | Rishabh     |
| 102            | Amit        |
| 103            | Neha        |

<br>

**Referance Table** Course Table
| CourseID (PK) | CourseName |
| ------------- | ---------- |
| C1            | DBMS       |
| C2            | OS         |

<br>

Student- Course Table
| StudentID (FK) | CourseID (FK) | Marks |
| -------------- | ------------- | ----- |
| 101            | C1            | 85    |
| 101            | C2            | 90    |
| 102            | C2            | 88    |
| 103            | C1            | 92    |



Now:

    StudentName depends only on StudentID (in STUDENT table)

    CourseName depends only on CourseID (in COURSE table)

    Marks depends on both StudentID and CourseID

## **3 Normal Form**

* It is already in Second Normal Form (2NF)

* There is no transitive dependency

<font color="red" >**Transitive Dependency** : A non- primary key attribute depends on another non-primary key attribute, instead of directly on the primary key   </font>

A → B and B → C  Then A → C (transitive dependency)

Example:

| StudentID | StudentName | DeptID | DeptName         |
| --------- | ----------- | ------ | ---------------- |
| 101       | Rishabh     | D1     | Computer Science |
| 102       | Amit        | D2     | Electronics      |
| 103       | Neha        | D1     | Computer Science |


**Functional Dependencies**

    StudentID → StudentName

    StudentID → DeptID

    DeptID → DeptName

    StudentID → DeptID → DeptName

  
**Base Table** Student Table
| StudentID (PK) | StudentName | DeptID (FK) |
| -------------- | ----------- | ----------- |
| 101            | Rishabh     | D1          |
| 102            | Amit        | D2          |
| 103            | Neha        | D1          |

<br>

**Referance Table** Department Table
| DeptID (PK) | DeptName         |
| ----------- | ---------------- |
| D1          | Computer Science |
| D2          | Electronics      |

<br>

Foreign Key Relationship: STUDENT.DeptID → DEPARTMENT.DeptID




## **BCNF Normal Form** (Boyce-Codd Normal Form)

* An advanced version of 3 Normal form
*  The table should be in the 3rd Normal Form
* It ensures every determinant left side of function dependency is Superkey or Candidate key
* X should be a super-key for every functional dependency (FD) X−>Y in a given relation.

      The normal forms become stricter as we move from 1NF → 2NF → 3NF → BCNF:

      1NF: Each field must hold atomic (indivisible) values.

      2NF: No partial dependency on a primary key.

      3NF: No transitive dependency on a primary key.

      BCNF: Every determinant must be a candidate key.
