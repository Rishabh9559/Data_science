# **DBMS Normal Forms**

## **1 Normal Form**

  * In 1 normal form each attribule have only one unit of context
  * Each field contains atomic (single) values
  * No repeating groups or multi-valued attrib

  example:
  | StudentID | StudentName | Subjects     |
| --------- | ----------- | ------------ |
| 101       | Rishabh     | DBMS, OS, CN |
| 102       | Amit        | OS, CN       |
| 103       | Neha        | DBMS         |


* Here in subject attribute/ column have more than one value in record so its not 1NF, we have to convert into single pr unit forms.

1NF solution: split table into Base table and Referance table

Base Table:
| StudentID (PK) | StudentName |
| -------------- | ----------- |
| 101            | Rishabh     |
| 102            | Amit        |
| 103            | Neha        |


Referance Table :
| StudentID (FK) | CourseName |
| -------------- | ---------- |
| 101            | DBMS       |
| 101            | OS         |
| 101            | CN         |
| 102            | OS         |
| 102            | CN         |
| 103            | DBMS       |








## **2 Normal Form**

  * Table or relation must be in 1 Normal Form.
  * All the non-prime attribute should be fully-functional dependent on Candidant key.

    <font color="red">**Candidant Key** : A candidate key in a database is a minimal set of one or more attributes (columns) that can uniquely identify each record (row) in a table </font>

  * There should be No Partial dependency in the relation.

Example :
| StudentID | CourseID | StudentName | CourseName | Marks |
| --------- | -------- | ----------- | ---------- | ----- |
| 101       | C1       | Rishabh     | DBMS       | 85    |
| 101       | C2       | Rishabh     | OS         | 90    |
| 102       | C2       | Amit        | OS         | 88    |
| 103       | C1       | Neha        | DBMS       | 92    |

<br>

Candidant key : StudentID + CourseID

<br>

**Functional Dependencies**:

  StudentID → StudentName

  CourseID → CourseName

  (StudentID, CourseID) → Marks

  So:

  StudentName depends only on StudentID

  CourseName depends only on CourseID

  These are partial dependencies → ❌ Not in 2NF

**Base Table** Student Table
| StudentID (PK) | StudentName |
| -------------- | ----------- |
| 101            | Rishabh     |
| 102            | Amit        |
| 103            | Neha        |

<br>

**Referance Table** Course Table
| CourseID (PK) | CourseName |
| ------------- | ---------- |
| C1            | DBMS       |
| C2            | OS         |

<br>

Student- Course Table
| StudentID (FK) | CourseID (FK) | Marks |
| -------------- | ------------- | ----- |
| 101            | C1            | 85    |
| 101            | C2            | 90    |
| 102            | C2            | 88    |
| 103            | C1            | 92    |



Now:

    StudentName depends only on StudentID (in STUDENT table)

    CourseName depends only on CourseID (in COURSE table)

    Marks depends on both StudentID and CourseID
