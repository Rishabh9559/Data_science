# **Unique Key and Foreign Key**

The main differences are that

**primary key** cannot contain NULL values, can only be one per table, and is typically the clustered index.

**A unique key** allows one NULL value, can have multiple per table, and is typically a non-clustered index.

Both keys enforce uniqueness in a column or set of columns.

## **Foreign Key**

A foreign key is a column or group of columns in a relational database table that provides a link between data in two tables. It is a column (or columns) that references a column (most often the primary key) of another table.

```
CREATE TABLE orders (
    order_id INT PRIMARY KEY,
    customer_id INT,
    FOREIGN KEY (customer_id) REFERENCES customers-table(customer_id)
);


```

#### Adding FOREIGN KEY to Existing Table

```
ALTER TABLE orders
ADD CONSTRAINT fk_customer
FOREIGN KEY (customer_id) REFERENCES customers-Table(customer_id);

```

### **FOREIGN KEY with CASCADE**

#### **ON DELETE CASCADE**

If parent row is deleted  child rows auto-delete.

```
FOREIGN KEY (customer_id)
REFERENCES customers(customer_id)
ON DELETE CASCADE;

```


#### **ON UPDATE CASCADE**

If PRIMARY KEY changes  child table updates automatically.

```
ON UPDATE CASCADE;
```

## **ORDER BY**  Sort the rows Ascending or Descending

* Order by a Single Column (Ascending - Default)

```
SELECT * FROM table-name ORDER BY column-table
```

* Order by a Single Column (Descending)

```
SELECT * FROM table-name ORDER BY column-name DESC;
```

* Order by Multiple Columns
```
SELECT * FROM table-name ORDER BY column-name ASC, column-name DESC;
```

* Order by Column Not Selected

```
SELECT studentID, CourseName FROM Student-table ORDER BY price DESC;
```

* ORDER BY with LIMIT (Top 3 highest fee courses)

```
SELECT student-name, CourseName, Price FROM student-table ORDER BY Price DESC LIMIT 4;
```
### **ORDER BY with GROUP BY**

* Count students per course, order by count
```
SELECT CourseID, COUNT(studentID) AS total_students
FROM Student
GROUP BY CourseID
ORDER BY total_students DESC;
```

* Minimum & Maximum duration per course, sorted by duration
```
SELECT CoursName, MIN(CoursDuration) AS min_dur, MAX(CoursDuration) AS max_dur
FROM Coures
GROUP BY CoursName
ORDER BY max_dur DESC;
```
