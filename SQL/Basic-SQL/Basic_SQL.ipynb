# **Basic SQL**

### **SQL** : Structured Query Language


SQL is a standardized programming language designed for managing and manipulating relational databases. It is used to perform various operations on data, such as:


**Retrieving data**: Extracting specific information from the database using queries.


**Manipulating data**: Inserting new data, updating existing data, and deleting data from tables.
Defining data

**structures**: Creating, modifying, and deleting database tables and their relationships.
Controlling data

**access**: Managing user permissions and security within the database.

## **SQL Data Types**

| Data Type        | Category        | Description                                               |
|------------------|------------------|-----------------------------------------------------------|
| INT              | Numeric          | Integer values (whole numbers).                          |
| BIGINT           | Numeric          | Large integer values.                                     |
| SMALLINT         | Numeric          | Small integer values.                                     |
| TINYINT          | Numeric          | Very small integer values (0255).                        |
| DECIMAL(p, s)    | Numeric          | Fixed-point values with precision and scale.              |
| NUMERIC(p, s)    | Numeric          | Same as DECIMAL, high precision.                          |
| FLOAT            | Numeric          | Floating-point number (approximate).                      |
| DOUBLE           | Numeric          | Double-precision floating-point.                          |
| CHAR(n)          | String           | Fixed-length character string.                            |
| VARCHAR(n)       | String           | Variable-length character string.                         |
| TEXT             | String           | Large text data.                                          |
| DATE             | Date/Time        | Stores calendar date (YYYY-MM-DD).                        |
| TIME             | Date/Time        | Stores time of day (HH:MM:SS).                            |
| DATETIME         | Date/Time        | Stores both date and time.                                |
| TIMESTAMP        | Date/Time        | Stores date & time with auto-update ability.              |
| YEAR             | Date/Time        | Stores year values.                                       |
| BOOLEAN          | Boolean          | TRUE or FALSE values.                                     |
| BINARY(n)        | Binary           | Fixed-length binary data.                                 |
| VARBINARY(n)     | Binary           | Variable-length binary data.                              |
| BLOB             | Binary           | Large binary objects (images, files).                     |
| ENUM             | Custom           | String list of allowed values.                            |
| SET              | Custom           | Multiple selected values from a predefined list.          |

# Keys


| Key Name           | Type              | Definition                                                                 |
|--------------------|-------------------|-----------------------------------------------------------------------------|
| PRIMARY KEY        | Uniqueness Key    | Uniquely identifies each row in a table; cannot contain NULL.              |
| FOREIGN KEY        | Relationship Key  | Links two tables; references PRIMARY KEY of another table.                 |
| CANDIDATE KEY      | Uniqueness Key    | Columns that qualify to become a PRIMARY KEY.                              |
| SUPER KEY          | Uniqueness Key    | Any set of columns that uniquely identify a row (including extra columns). |
| COMPOSITE KEY      | Uniqueness Key    | A primary/unique key made of 2 or more columns.                            |
| UNIQUE KEY         | Uniqueness Key    | Ensures values in a column are unique, but allows one NULL.                |
| ALTERNATE KEY      | Uniqueness Key    | Candidate keys that are not selected as PRIMARY KEY.                       |
| CHECK CONSTRAINT   | Validation Key    | Ensures data meets specific conditions.                                    |
| NOT NULL           | Validation Key    | Ensures a column cannot store NULL values.                                 |
| DEFAULT            | Validation Key    | Sets a default value for a column when no value is provided.               |
| INDEX              | Performance Key   | Improves search and query performance on columns.                          |
| FULLTEXT INDEX     | Search Key        | Used for full-text search in large text fields.                            |
| UNIQUE INDEX       | Uniqueness Key    | Ensures indexed column values are unique.                                  |
