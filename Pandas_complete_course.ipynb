## **DataFrame**

Dat sets in Pandas are usually multi-dimensional tables, called DataFrames.

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
# DataFrame:

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Alice"],
    "Age": [25, 30, 35, np.nan, 29, 25],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "HR"],
    "Salary": [50000, 60000, 70000, 62000, np.nan, 50000]
}

data
```

```python
# DataFrame
df = pd.DataFrame(data)
df
```

### **loc and iloc**

* **loc** (Label-based): Access rows/columns by name or label.
  * **Example**: df.loc[0:5, 'ColumnName'] (Includes the row with label 5).

* **iloc** (Integer-based): Access rows/columns by numerical index positions.
  * **Example**: df.iloc[0:5, 2] (Excludes the row at position 5).
* **Slicing Behavior**: **loc** includes both start and stop, **iloc** includes start but excludes stop.

* **Performance**: **iloc** is generally faster as it works directly with memory positions.

```python
# loc and iloc

#loc
df.loc[1:4]
```

```python
# loc- loc[index, [column name]]
df.loc[0:3 , ["Age", "Department"]]
```

```python
# loc -> retrive by index value
print(df.loc[0])
print("-------------------")
print(df.loc[2:4])

```

```python
# iloc -> only retrive integer index
print(df.iloc[0])

print("-----------------")

print(df.iloc[1:4])
```

### **`pandas.DataFrame.describe()`**

The `pandas.DataFrame.describe()` method generates **descriptive statistics** that summarize the **central tendency**, **dispersion**, and **shape** of a dataset’s distribution.  

By default, it analyzes **only numeric columns** and **excludes missing (NaN) values**.



#### Key Statistics Provided

#### **For Numeric Data (default)**
- **count**: Number of non-null observations  
- **mean**: The average value  
- **std**: The standard deviation (measure of spread)  
- **min**: The minimum value  
- **25%**: The 25th percentile (first quartile)  
- **50%**: The 50th percentile (median)  
- **75%**: The 75th percentile (third quartile)  
- **max**: The maximum value  

#### **For Object / Categorical Data**  
*(e.g., strings or timestamps, when specified)*
- **count**: Number of non-null observations  
- **unique**: Number of unique values  
- **top**: Most frequently occurring value  
- **freq**: Frequency (count) of the most frequent value  


```python
# describe()

df.describe()
```
