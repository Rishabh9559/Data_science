# **Percentage:**
A percentage is a way of expressing a number as a fraction of 100. It's often used to represent a proportion or a part of a whole.
For example, if you have 50 apples out of a total of 200 apples, the percentage of apples you have is (50 / 200) * 100% = 25%.
The symbol for percentage is "%".

# **Percentile:**
A percentile is a measure used in statistics indicating the value below which a given percentage of observations in a group of observations fall. For example, the 20th percentile is the value below which 20% of the observations may be found.
Percentiles are often used in standardized testing to indicate how a student's score compares to the scores of other students. If you score in the 90th percentile on a test, it means you scored higher than 90% of the other test-takers.
Percentiles are also used to understand the distribution of data. For instance, the median of a dataset is the 50th percentile.

**Key Differences:**

*   **What they represent:** Percentage represents a part of a whole or a proportion relative to 100. Percentile represents a position within a dataset relative to others in the dataset.
*   **Calculation:** Percentage is calculated as (part / whole) * 100. Percentile is calculated by finding the value below which a certain percentage of the data falls.
*   **Use cases:** Percentages are used for expressing proportions, discounts, taxes, etc. Percentiles are used for comparing individual scores or values to a distribution, understanding data spread, and standardized testing.

# **IQR**  - Interquartile Range

is a measure of statistical dispersion, specifically representing the spread of the middle 50% of a data set. It is calculated by subtracting the first quartile (Q1) from the third quartile (Q3).

# IQR = Q3 - Q1

- Q3 : half of second half
- Q1 : half of first half

Outliners = ( Q1 + IQR , Q3 + 1.5*IQR )
