**Introduction of Statistics**

Statistics is the science concerned with developing and studying methods for collecting, analyzing, interpreting and presenting empirical data

Two main types: descriptive statistics and inferential statistics

**Descriptive statistics:**  provide a concise summary of the main characteristics of a dataset, using numbers and visuals to describe key features like central tendency (mean, median, mode) and dispersion (variance, standard deviation)

**Inferential Statistics** :uses data from a sample to make conclusions or inferences about a larger population.

 It's about drawing generalizations and making predictions about a whole group based on a smaller subset

image.png

**Population VS Sample**

Population - refers to the entire group of individuals or objects

Sample - sample is a subset of that population that is selected for analysis

**Data:** it is simplest form, is a collection of facts, figures, or symbols that can be processed or analyzed to extract useful information

**Types of Data**
- **Qualitative data (descriptive):** Qualitative data describes the characteristics or qualities of something, rather than numerical values.

  Norminal and Ordinal Data

  - Example- interview transcripts, diary entries, field notes, open-ended survey responses, and visual data like photographs and videos.

- **Quantitative (numerical):** Quantitative or numerical data refers to information expressed in numbers, allowing for mathematical analysis and statistical evaluation.
  -  Examples include height, weight, income, test scores, and sales figures.

```python
# generating own data

import pandas as pd
import numpy as np

np.random.seed(0)
num_samples=1000
colors=['Red','Green','Grey','Blue','Purple','Black','White','Yellow']
nominal_data=np.random.choice(colors,num_samples)
nominal_data

```

```python
## data frame define
dataframe=pd.DataFrame(nominal_data,columns=['colors'])
dataframe.head(10)
```

```python
# genetaed new column in dataframe

satisfaction_level=['low','median','high']
satisfaction_data=np.random.choice(satisfaction_level,size=num_samples)
dataframe['satisfaction']=satisfaction_data
dataframe.head(10)
```

```python
# creating tempture column
np.random.seed(0)
temp_data=np.random.randint(0,100,num_samples)
dataframe['Tempeture']=temp_data
dataframe.tail()
```

```python
# range data income column in perticuar range

np.random.seed(0)
income_Data=np.random.randint(50000,200000,num_samples)
dataframe['Income']=income_Data
dataframe.head()
```

```python
# Histograph plot
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(dataframe['Income'],color='pink',alpha=0.7,edgecolor='red',bins=30)
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Income Distribution')
plt.show()
```
