**Statistical Test**

A statistical test is a method used to assess evidence for or against a hypothesis about a population parameter based on sample data.

A statistical test is a mathematical method used to determine whether there is enough evidence in a sample of data to support a specific claim (called a hypothesis) about a population.

image.png

Key Concepts:

- Hypotheses:
  1. Null Hypothesis (H₀): A default assumption that there is no effect or no difference.

  2. Alternative Hypothesis (H₁ or Ha): Suggests that there is an effect or a difference.

- Purpose:
To assess whether observed data significantly deviates from what we'd expect under the null hypothesis.

- Test Statistic:
A value calculated from the sample data, which is used to determine whether to reject H₀.

- P-value:
The probability of obtaining a result at least as extreme as the one observed, assuming H₀ is true.

- Significance Level (α):
A threshold (commonly 0.05). If p-value < α, reject H₀.

Types of Statistical Tests:

1. Parametric Tests:
These tests assume data follows a specific distribution (e.g., normal distribution) and utilize population parameters (like means and standard deviations) in their calculations. Examples include t-tests, z-tests, and ANOVA.

2. Non-parametric Tests:
These tests do not make assumptions about the data distribution and are used when data is not normally distributed or when dealing with ordinal or ranked data. Examples include the Mann-Whitney U test, the Kruskal-Wallis test, and the Wilcoxon signed-rank test.

Examples:

- Testing the effectiveness of a new drug:
A t-test could be used to compare the recovery rates of patients treated with the new drug versus a placebo.

- Determining if there is a relationship between smoking and lung cancer:
A chi-square test could be used to analyze the association between smoking status and the presence or absence of lung cancer.

- Comparing the average test scores of students in different schools:
ANOVA could be used to see if there are significant differences in test scores between multiple schools.

**One-tailed vs Two-tailed Tests are types of hypothesis tests**

### **Two-tailed Test**

A **two-tailed test** checks **both directions** — whether a sample is significantly **greater than or less than** the population value.

#### Used When:

You're testing **for any significant difference**, regardless of direction.

#### Example:

> **Hypothesis:** Does a new medicine affect blood pressure differently than the old one?

* H₀: µ = 120 (mean blood pressure is 120)
* H₁: µ ≠ 120 (it could be **higher or lower**)

You’re interested in **any change**, not just an increase or decrease.

---

### **One-tailed Test**

A **one-tailed test** checks in **one direction only** — either for a significant **increase** or **decrease**.

#### Used When:

You're testing for an effect in a **specific direction**.

#### Example 1 (Right-tailed):

> **Hypothesis:** Does a new fertilizer **increase** crop yield?

* H₀: µ ≤ 50 kg (no increase)
* H₁: µ > 50 kg (**increase** only)

#### Example 2 (Left-tailed):

> **Hypothesis:** Does a new policy **reduce** average waiting time?

* H₀: µ ≥ 30 min
* H₁: µ < 30 min (**decrease** only)

---

## ⚖️ **Comparison Summary:**

| Feature          | One-tailed Test                       | Two-tailed Test                |
| ---------------- | ------------------------------------- | ------------------------------ |
| Direction tested | One (left or right)                   | Both (left and right)          |
| H₁ form          | µ > value or µ < value                | µ ≠ value                      |
| Sensitivity      | More powerful if direction is correct | More conservative              |
| Example          | "Is it **more than** 50?"             | "Is it **different from** 50?" |
